```python
# Accept scores from user and save scores in a list. 
# Then calculate the sum of all scores and their average score.
# Finally, print out all scores and their average

total = 0
answer = ''
scores = []

while answer.lower() != 'n' :
    num = int(input("Enter a score: "))
    scores.append(num)
    total += num
    answer = input("Do you want to continue (y/n)? ")
    
avg = total / len(scores)
print(scores)
print(f"avg score is {avg:.2f}")
```

### Exercise 3
The following function will determine if a number is a prime number or not, but it contains bugs. Fix the bugs and make sure it works. 

```python
#solution

def prime(numb) :
    for y in range(2, numb) : # 2, 3, 4, 5, 6, 7, 8
        if numb % y == 0 :
            msg = str(numb) + ' is not a prime number'
            break
    else:
        msg = str(numb) + ' is a prime number'   
        
    return msg    
```

```python
# test prime function

print(prime(2))
print(prime(20))
print(prime(11))
print(prime(100))
```

```python
# main program

numb = int(input("Enter a number: "))
for x in range (2, numb) : 
    print(prime(x))              
        
print("Done!")        
```

#### Now, rewrite prime function, using a while-else statement

```python
def prime_while(numb) :
    y = 2
    while y < numb :
        if numb % y == 0 :
            msg = str(numb) + ' is not a prime number'
            break
        y += 1
    else :
        msg = str(numb) + ' is a prime number' 
    
    return msg

```

```python
print(prime_while(2))
print(prime_while(20))
print(prime_while(11))
print(prime_while(100))
```

### Nested loops
loop inside another loop

```python
for outer in ["first", "second", "third"] :
    print(outer)
    
    for inner in range(3) :
        print(inner+1)
```

```python
answer = 'y'
while answer != 'n' :
    index = int(input("Enter a number: "))
    for inner in range(index) :
        print(inner+1)
        
    answer = input("Continue? (y/n)")  
    print("*" * 20)

print("Done!")      
```

```python
answer = 'y'
questions = ["your id", "your password"]
    
while answer != 'n' :    
    no_qs = len(questions)
    for inner in range(no_qs) :
        msg = "What is " + questions[inner] + "? "
        answer = input(msg)
    
    answer = input("What is your additional security question? ")
    questions.append(answer)
    answer = input("Continue? (y/n)")
    print("*" * 20)
    
print("Done!")    
```
